```python
#Bibliotheken importieren
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

#Iris Dataset von scikit-learn lesen und einen DataFrame erstelen
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
print(data)

#labels in einer Variable reinspeichern)
labels = pd.Categorical.from_codes(iris.target, iris.target_names)
#print(labels)

#
def GenerateDTreeFrom_nD_Data(Data, Labels, Metric="gini"):
    dtree = DecisionTreeClassifier(criterion=Metric)
    
    dtree.fit(Data, Labels)
    
    return dtree
gdtree = GenerateDTreeFrom_nD_Data(data, labels, Metric="gini")
gdtree
```

```python
from IPython.display import Image  
#import six 
from io import StringIO  
from sklearn.tree import export_graphviz
import pydot 

features = list(data.columns[:])
features
 
```

```python
dot_data = StringIO()  
export_graphviz(gdtree, 
                out_file=dot_data,
                feature_names=features,
                filled=True,
                rounded=True)

graph = pydot.graph_from_dot_data(dot_data.getvalue())  
Image(graph[0].create_png())
```
